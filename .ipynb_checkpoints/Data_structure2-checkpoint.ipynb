```python
#5.Describe the key features of  sets and provide examples of their uses in python.
#         Sets in Python are unordered collections of unique elements. They are commonly used for membership testing, eliminating duplicate entries, and performing mathematical set operations like union, intersection, and difference. Here are the key features of sets and examples of their uses:

#         Key Features of Sets
#         Unordered:
#         Sets do not maintain any order of elements. Therefore, you cannot access elements via indexing.

        my_set = {1, 2, 3}
        # print(my_set[0])  # This will raise an error
        
        #Unique Elements:
        #Sets automatically eliminate duplicate entries.
        
        my_set = {1, 2, 2, 3, 4, 4}
        print(my_set)  # Output: {1, 2, 3, 4}
#         Mutable:
#         Sets are mutable, meaning you can add or remove elements. However, the elements themselves must be immutable (e.g., strings, numbers, tuples).

#         No Indexing:
#         Since sets are unordered, you cannot access elements by index.

#         Set Operations:
#         Sets support mathematical operations like union, intersection, difference, and symmetric difference.

#         Creating Sets
#         Using curly braces {}:

        my_set = {1, 2, 3}
       
        #Common Set Methods and Operations
        #Adding Elements:
        
        my_set = {1, 2, 3}
        my_set.add(4)
        print(my_set)  # Output: {1, 2, 3, 4}
       # Removing Elements:

        my_set.remove(3)  # Raises a KeyError if the element is not found
        print(my_set)  # Output: {1, 2, 4}

        my_set.discard(2)  # Does not raise an error if the element is not found
        print(my_set)  # Output: {1, 4}
        
        #Set Operations:

        #Union: Combines elements from both sets.

        set1 = {1, 2, 3}
        set2 = {3, 4, 5}
        union_set = set1 | set2
        print(union_set)  # Output: {1, 2, 3, 4, 5}
        
        #Intersection: Returns elements common to both sets.

        intersection_set = set1 & set2
        print(intersection_set)  # Output: {3}
       
    #Difference: Returns elements in the first set but not in the second.

        difference_set = set1 - set2
        print(difference_set)  # Output: {1, 2}
       # Symmetric Difference: Returns elements in either set, but not in both.

        sym_diff_set = set1 ^ set2
        print(sym_diff_set)  # Output: {1, 2, 4, 5}
        
        #Checking Membership:

        print(1 in my_set)  # Output: True
        print(5 in my_set)  # Output: False
        
        #Examples of Set Uses
        #Removing Duplicates:

        my_list = [1, 2, 2, 3, 4, 4, 5]
        my_set = set(my_list)
        unique_list = list(my_set)
        print(unique_list)  # Output: [1, 2, 3, 4, 5]
       
        #Membership Testing:

        allowed_users = {"alice", "bob", "carol"}
        user = "dave"
        if user in allowed_users:
            print("Access granted")
        else:
            print("Access denied")
        # Output: Access denied
       
        #Set Operations:

        set_a = {1, 2, 3, 4}
        set_b = {3, 4, 5, 6}

        union = set_a | set_b
        print("Union:", union)  # Output: Union: {1, 2, 3, 4, 5, 6}

        intersection = set_a & set_b
        print("Intersection:", intersection)  # Output: Intersection: {3, 4}

        difference = set_a - set_b
        print("Difference:", difference)  # Output: Difference: {1, 2}

        sym_diff = set_a ^ set_b
        print("Symmetric Difference:", sym_diff)  # Output: Symmetric Difference: {1, 2, 5, 6}
        
        #Eliminating Duplicates from a List:

        numbers = [1, 2, 3, 4, 5, 5, 6, 7, 8, 8, 9]
        unique_numbers = list(set(numbers))
        print(unique_numbers)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
        #Sets are highly useful for scenarios where the uniqueness of elements is essential or when performing mathematical set operations.
        
#6.  discuss the use cases of tuples and sets in python programming.     
#           Use Cases of Tuples in Python Programming
          # Fixed Collections of Items:
          # Tuples are used to store collections of items that should not change throughout the program. For example, the coordinates of a point, RGB color values, or days of the week.

        coordinates = (10, 20)
        colors = (255, 255, 255)  # RGB color
        days_of_week = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
        
        #Immutable Records:
        #Tuples are suitable for representing records that should not be altered, such as database rows or configuration settings.

    
        person = ("Alice", 30, "Engineer")
        
        #Dictionary Keys:
        #Tuples can be used as keys in dictionaries because they are immutable.

        #Performance Optimization:
        #Since tuples are immutable, they can be used to optimize performance in scenarios where the collection of items should not change, leading to more efficient memory usage.

        data = (1, 2, 3, 4, 5)
        
        #Packing and Unpacking:
        #Tuples facilitate packing and unpacking operations, simplifying the assignment of multiple variables.

        person = ("Alice", 30, "Engineer")
        name, age, profession = person
        
        # Use Cases of Sets in Python Programming
        # Membership Testing:
        # Sets are highly efficient for testing whether an element is part of a collection, making them useful for scenarios like user permissions or blacklist checks.

        allowed_users = {"alice", "bob", "carol"}
        if user in allowed_users:
            print("Access granted")
        
        # Eliminating Duplicates:
        # Sets automatically remove duplicate elements, making them ideal for filtering unique items from a collection.

        items = [1, 2, 2, 3, 4, 4, 5]
        unique_items = list(set(items))
       
        # Mathematical Set Operations:
        # Sets support operations like union, intersection, difference, and symmetric difference, useful for tasks like finding common or distinct elements in datasets.

        set_a = {1, 2, 3}
        set_b = {2, 3, 4}

        union = set_a | set_b  # {1, 2, 3, 4}
        intersection = set_a & set_b  # {2, 3}
        difference = set_a - set_b  # {1}
        sym_diff = set_a ^ set_b  # {1, 4}
        
        email_addresses = {"user1@example.com", "user2@example.com"}
        if new_email in email_addresses:
            print("Email already exists")
       # Both tuples and sets provide unique advantages in Python programming, making them suitable for a wide range of applications that benefit from immutability, uniqueness, and efficient data operations.

#7.Describe how to add , modify , delete items in a dictionary  with examples.

#            Adding Items to a Dictionary
#         Adding a New Key-Value Pair:

        my_dict = {"name": "Alice", "age": 25}
        my_dict["city"] = "New York"
        print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
        
        #Using update() Method:

        my_dict.update({"email": "alice@example.com"})
        print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York', 'email': 'alice@example.com'}
        
        #Modifying Items in a Dictionary
        #Modifying an Existing Value:

        my_dict = {"name": "Alice", "age": 25, "city": "New York"}
        my_dict["age"] = 26
        print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}
       
    #Using update() Method:

        my_dict.update({"city": "Los Angeles"})
        print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'}
        
        #Deleting Items from a Dictionary
        #Using del Statement:

        my_dict = {"name": "Alice", "age": 26, "city": "Los Angeles"}
        del my_dict["age"]
        print(my_dict)  # Output: {'name': 'Alice', 'city': 'Los Angeles'}
        
        #Using pop() Method:

        city = my_dict.pop("city")
        print(city)     # Output: 'Los Angeles'
        print(my_dict)  # Output: {'name': 'Alice'}
       
        #Using popitem() Method
        
        my_dict = {"name": "Alice", "age": 26, "city": "Los Angeles"}
        item = my_dict.popitem()
        print(item)     # Output: ('city', 'Los Angeles')
        print(my_dict)  # Output: {'name': 'Alice', 'age': 26}
        
        #Using clear() Method:

        my_dict.clear()
        print(my_dict)  # Output: {}
        
        
        #Adding Items
        
        # Initial dictionary
        person = {"name": "John", "age": 30}

        # Add a new key-value pair
        person["city"] = "New York"
        print(person)  # Output: {'name': 'John', 'age': 30, 'city': 'New York'}

        # Using update() method
        person.update({"email": "john@example.com"})
        print(person)  # Output: {'name': 'John', 'age': 30, 'city': 'New York', 'email': 'john@example.com'}
        
        #Modifying Items

        # Modify an existing key-value pair
        person["age"] = 31
        print(person)  # Output: {'name': 'John', 'age': 31, 'city': 'New York', 'email': 'john@example.com'}

        # Using update() method to modify a value
        person.update({"city": "Los Angeles"})
        print(person)  # Output: {'name': 'John', 'age': 31, 'city': 'Los Angeles', 'email': 'john@example.com'}
       
        #Deleting Items
       
        # Using del statement
        del person["email"]
        print(person)  # Output: {'name': 'John', 'age': 31, 'city': 'Los Angeles'}

        # Using pop() method
        
        age = person.pop("age")
        print(age)  # Output: 31
        print(person)  # Output: {'name': 'John', 'city': 'Los Angeles'}

        # Using clear() method
        person.clear()
        print(person)  # Output: {}
        #These examples demonstrate how to add, modify, and delete items in a dictionary, showcasing the flexibility and power of this data structure in Python.

#8.Discuss the importance of dictionary keys being immutable and provide examples.
#            In Python, dictionary keys must be immutable, which means that the keys cannot be changed after they are created. This immutability requirement is crucial for maintaining the integrity and functionality of the dictionary. Here's a detailed discussion of the importance of dictionary keys being immutable, along with examples:

#         Importance of Dictionary Keys Being Immutable
#         Hashability:

#         Dictionary keys must be hashable, meaning they must have a constant hash value during their lifetime. Immutable types, such as strings, numbers, and tuples, have fixed hash values.
#         The hash value is used to quickly compare dictionary keys during lookups, inserts, updates, and deletions. If the key could change, its hash value would also change, making it impossible to retrieve the associated value correctly.
#         Data Integrity:

#         Immutable keys ensure that the key-value pairs in the dictionary remain consistent and reliable. If keys were mutable, modifying a key would corrupt the dictionary, as the key's new value would not match the original hash or position in the hash table.
       # Performance:

#         Dictionary operations, such as lookups, are designed to be O(1) on average, thanks to the constant-time complexity of hashing. Mutable keys could disrupt this efficiency, leading to slower operations and increased complexity.
#         Security:

#         Immutable keys reduce the risk of accidental or malicious alterations to dictionary keys. This helps prevent unexpected behavior or errors in programs.
#         Examples of Immutable and Mutable Types
#         Immutable Types (Valid Dictionary Keys):
#         String:

        my_dict = {"name": "Alice", "age": 30}
        print(my_dict["name"])  # Output: Alice
       
        #Number:

    
        my_dict = {1: "one", 2: "two", 3: "three"}
        print(my_dict[2])  # Output: two

    
        my_dict = {(1, 2): "point1", (3, 4): "point2"}
        print(my_dict[(1, 2)])  # Output: point1
        
        #Mutable Types (Invalid Dictionary Keys):
        #List:

        my_dict = {[1, 2, 3]: "list"}  # Raises TypeError: unhashable type: 'list'
        Dictionary:

        my_dict = {{'a': 1}: "dict"}  # Raises TypeError: unhashable type: 'dict'
       
        #Set:
        
        my_dict = {{1, 2, 3}: "set"}  # Raises TypeError: unhashable type: 'set'
        # Practical Examples
        # Using Immutable Keys (Correct Usage):
      
        # Using strings as keys
        person = {"name": "Alice", "age": 30, "city": "New York"}
        print(person["age"])  # Output: 30

        # Using numbers as keys
        squares = {1: 1, 2: 4, 3: 9, 4: 16}
        print(squares[3])  # Output: 9

        # Using tuples as keys
        points = {(1, 2): "A", (3, 4): "B", (5, 6): "C"}
        print(points[(3, 4)])  # Output: B
       
     
        #In summary, the requirement for dictionary keys to be immutable is essential for maintaining the performance, integrity, and reliability of dictionaries in Python. Immutable keys ensure that the hash values remain constant, which is critical for efficient lookups and other dictionary operations.

```
