```python
#1.Discuss string slicing and provide examples.
#         String slicing is a technique in Python used to extract a portion of a string using a range of indices. This allows you to obtain substrings without modifying the original string.

#         The basic syntax for string slicing is:

#         python
#         Copy code
#         string[start:end:step]
#         start: The starting index (inclusive).
#         end: The ending index (exclusive).
#         step: The step size (default is 1).
#         Here are some examples to illustrate string slicing:

#         Basic Slicing:
         text = "Hello, World!"
           print(text[0:5])  # Output: Hello
        #In this example, text[0:5] extracts the substring from index 0 to index 4 (not including index 5).

        #Omitting start or end:

        print(text[:5])   # Output: Hello
        print(text[7:])   # Output: World!
        #If start is omitted, slicing starts from the beginning of the string. If end is omitted, slicing goes to the end of the string.

       # Negative Indices:

        print(text[-6:])  # Output: World!
        print(text[-6:-1])  # Output: World
       # Negative indices count from the end of the string. text[-6:] extracts the last 6 characters, and text[-6:-1] extracts from the 6th last character to the second last character.

       # Using step:

        print(text[::2])  # Output: Hlo ol!
        print(text[1::2])  # Output: el,Wrd
      #  The step parameter determines the interval between indices. text[::2] takes every second character from the string, starting at index 0. text[1::2] starts at index 1 and takes every second character.

        #Reversing a String:

       
        print(text[::-1])  # Output: !dlroW ,olleH
        #Using a step of -1 reverses the string.

        #Combining Slicing with Other Operations:

       
        new_text = text[:5] + text[7:]
        print(new_text)  # Output: HelloWorld!
        #Here, we combined slicing with concatenation to remove a part of the string (the comma and the space).

        #String slicing is powerful and can be used in various ways to manipulate and analyze strings efficiently.

#2.Explain the key features of lists in python.
   # Lists in Python are versatile, mutable sequences that can store a collection of items. These items can be of any data type, including other lists. Here are some key features of lists in Python:

        #a. Ordered
        #Lists maintain the order of elements. The order in which you insert elements is preserved, and you can access elements by their index.


        my_list = [1, 2, 3, 4]
        print(my_list[0])  # Output: 1
        print(my_list[3])  # Output: 4
        #b. Mutable
       # Lists are mutable, meaning you can change their content without changing their identity.

        my_list = [1, 2, 3]
        my_list[1] = 4
        print(my_list)  # Output: [1, 4, 3]
        #c. Heterogeneous Elements
        A single list can contain elements of different data types.

        my_list = [1, "Hello", 3.14, [5, 6]]
        print(my_list)  # Output: [1, 'Hello', 3.14, [5, 6]]
        #d. Dynamic Size
        #Lists can grow and shrink in size as needed. You can add or remove elements after the list is created.

        
        my_list = [1, 2, 3]
        my_list.append(4)
        print(my_list)  # Output: [1, 2, 3, 4]
        my_list.remove(2)
        print(my_list)  # Output: [1, 3, 4]
        #e. Comprehensions
        #Python provides list comprehensions, a concise way to create lists.

        squares = [x**2 for x in range(5)]
        print(squares)  # Output: [0, 1, 4, 9, 16]
        #f. Built-in Methods
       # Lists come with many useful built-in methods. Here are a few examples:

        # append(): Adds an element to the end of the list.
        # extend(): Extends the list by appending elements from an iterable.
        # insert(): Inserts an element at a specified position.
        # remove(): Removes the first occurrence of a specified element.
        # pop(): Removes and returns the element at the specified position (default is the last element).
        # clear(): Removes all elements from the list.
        # index(): Returns the index of the first occurrence of a specified element.
        # count(): Returns the number of occurrences of a specified element.
        # sort(): Sorts the elements of the list in place.
        # reverse(): Reverses the elements of the list in place.
        
        my_list = [3, 1, 2]
        my_list.sort()
        print(my_list)  # Output: [1, 2, 3]
       # g. Slicing
       # Similar to strings, lists can be sliced to create sublists.

        my_list = [1, 2, 3, 4, 5]
        print(my_list[1:4])  # Output: [2, 3, 4]
        print(my_list[:3])   # Output: [1, 2, 3]
        print(my_list[3:])   # Output: [4, 5]
        print(my_list[::-1]) # Output: [5, 4, 3, 2, 1]
       # h. Iterability
       # Lists are iterable, allowing you to loop through elements using loops.

        my_list = [1, 2, 3]
        for item in my_list:
            print(item)
        # Output:
        # 1
        # 2
        # 3
        #i. Nested Lists
        #Lists can contain other lists as elements, creating nested or multidimensional lists.

        nested_list = [[1, 2, 3], [4, 5, 6]]
        print(nested_list[0][1])  # Output: 2
        print(nested_list[1][2])  # Output: 6
        #These features make lists a powerful and flexible data structure in Python.
        
#3.describe how to access ,delete and modify elements in a list with examples.

        
        #Accessing Elements in a List
        #You can access elements in a list using indexing, which starts at 0 for the first element.

        #Single Element Access:

        my_list = [10, 20, 30, 40, 50]
        print(my_list[0])  # Output: 10
        print(my_list[3])  # Output: 40
        
        #Negative Indexing:
        print(my_list[-1])  # Output: 50 (last element)
        print(my_list[-2])  # Output: 40 (second last element)
        
        #Slicing:
        print(my_list[1:4])  # Output: [20, 30, 40]
        print(my_list[:3])   # Output: [10, 20, 30]
        print(my_list[3:])   # Output: [40, 50]
        print(my_list[::-1]) # Output: [50, 40, 30, 20, 10]
        # Deleting Elements in a List
        # Using del Statement:

        
        my_list = [10, 20, 30, 40, 50]
        del my_list[1]  # Deletes the element at index 1
        print(my_list)  # Output: [10, 30, 40, 50]
        
        #Using remove() Method:

        my_list.remove(30)  # Removes the first occurrence of 30
        print(my_list)  # Output: [10, 40, 50]
        Using pop() Method:

        my_list.pop()  # Removes and returns the last element
        print(my_list)  # Output: [10, 40]
        my_list.pop(0)  # Removes and returns the element at index 0
        print(my_list)  # Output: [40]
        
        #Using clear() Method:

        my_list.clear()  # Removes all elements from the list
        print(my_list)  # Output: []
        
        #Modifying Elements in a List
       # Assigning a New Value:

        my_list = [10, 20, 30, 40, 50]
        my_list[2] = 35  # Changes the element at index 2
        print(my_list)  # Output: [10, 20, 35, 40, 50]
       # Modifying Multiple Elements:

        my_list[1:3] = [25, 35]  # Changes elements at index 1 and 2
        print(my_list)  # Output: [10, 25, 35, 40, 50]
        #Using List Comprehension:

        my_list = [10, 20, 30, 40, 50]
        my_list = [x*2 for x in my_list]  # Doubles each element in the list
        print(my_list)  # Output: [20, 40, 60, 80, 100]
        #Using Methods:

        #Appending an Element:
        
        my_list.append(60)  # Adds 60 to the end of the list
        print(my_list)  # Output: [10, 20, 30, 40, 50, 60]
        
        #Inserting an Element:
    
        my_list.insert(1, 15)  # Inserts 15 at index 1
        print(my_list)  # Output: [10, 15, 20, 30, 40, 50, 60]
        
        #Extending a List:
        
        my_list.extend([70, 80])  # Adds elements from another list to the end
        print(my_list)  # Output: [10, 15, 20, 30, 40, 50, 60, 70, 80]
        
        #Sorting a List:
        
        my_list.sort()
        print(my_list)  # Output: [10, 15, 20, 30, 40, 50, 60, 70, 80]
        
        #Reversing a List:
        
        my_list.reverse()
        print(my_list)  # Output: [80, 70, 60, 50, 40, 30, 20, 15, 10]

#4.compare and contrast tuples and lists with examples.
  
#         Tuples and lists are both sequence data types in Python, but they have some key differences and similarities. Here’s a detailed comparison:

#         Similarities
#         Ordered:
#         Both tuples and lists maintain the order of elements.

        
        my_list = [1, 2, 3]
        my_tuple = (1, 2, 3)
        print(my_list[0])  # Output: 1
        print(my_tuple[0]) # Output: 1
        
        #Indexing and Slicing:
       # Both support indexing and slicing.

        print(my_list[1:3])  # Output: [2, 3]
        print(my_tuple[1:3]) # Output: (2, 3)
        
        #Heterogeneous Elements:
        #Both can store elements of different data types.

        my_list = [1, "Hello", 3.14]
        my_tuple = (1, "Hello", 3.14)
        
        #Differences
       # Mutability:

        #Lists are mutable. You can change, add, or remove elements.
       # Tuples are immutable. Once a tuple is created, it cannot be modified.
        
        # List modification
        my_list = [1, 2, 3]
        my_list[1] = 4
        print(my_list)  # Output: [1, 4, 3]

        # Tuple modification (will raise an error)
        my_tuple = (1, 2, 3)
        # my_tuple[1] = 4  # TypeError: 'tuple' object does not support item assignment
        #Syntax:

        # Lists are created using square brackets [].
        # Tuples are created using parentheses ().
        
        my_list = [1, 2, 3]
        my_tuple = (1, 2, 3)
       # Performance:
        #Tuples can be more performant than lists, especially when dealing with large collections of data, because they are immutable.

       # Use Cases:

#         Lists are typically used for collections of items that can change, such as a list of users, items in a shopping cart, etc.
#         Tuples are often used for fixed collections of items, such as the coordinates of a point (x, y), days of the week, etc.
#         Methods:

#         Lists have several built-in methods like append(), extend(), insert(), remove(), pop(), clear(), sort(), reverse().
#         Tuples have only two built-in methods: count() and index().
#       
        my_list = [1, 2, 2, 3]
        my_list.append(4)
        print(my_list)  # Output: [1, 2, 2, 3, 4]

        my_tuple = (1, 2, 2, 3)
        print(my_tuple.count(2))  # Output: 2
        print(my_tuple.index(3))  # Output: 3
       
        #Length:
        #Both use the len() function to get the number of elements.

        
        print(len(my_list))  # Output: 5
        print(len(my_tuple)) # Output: 4
        # Example Usage:
        # List Example
        
        # Creating a list of fruits
        fruits = ['apple', 'banana', 'cherry']
        # Adding an item to the list
        fruits.append('orange')
        print(fruits)  # Output: ['apple', 'banana', 'cherry', 'orange']
        # Modifying an item in the list
        fruits[1] = 'blueberry'
        print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'orange']
        #Tuple Example
        
        # Creating a tuple of coordinates
        coordinates = (10, 20)
        # Accessing elements
        print(coordinates[0])  # Output: 10
        print(coordinates[1])  # Output: 20
        # Trying to modify (will raise an error)
        # coordinates[1] = 30  # TypeError: 'tuple' object does not support item assignment







```
